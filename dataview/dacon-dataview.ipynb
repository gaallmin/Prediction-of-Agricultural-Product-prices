# train -> train.csv

```python
import pandas as pd

# train - train.csv
df = pd.read_csv(r"data\train\train.csv", encoding = "utf-8")

columns = ['품목명' , '품종명', '등급']
for column in columns:
        unique_values = df[column].unique()
        print(f"컬럼: {column}")
        print(f"고유값: {unique_values}")
        print("-" * 40)
```

```python
if '품목명' in columns and '품종명' in columns:
        unique_combinations = df[['품목명', '품종명']].drop_duplicates()
        print("unique pair (품목명, 품종명):")
        print(unique_combinations)
```

# train -> meta -> TRAIN_산지공판장.csv

```python
df = pd.read_csv(r"data\train\meta\TRAIN_산지공판장_2018-2021.csv", encoding = "utf-8")

columns = ['공판장코드' , '공판장명', '품목코드', '품목명', '품종코드', '품종명', '등급코드', '등급명']
for column in columns:
        unique_values = df[column].unique()
        print(f"컬럼: {column}")
        print(f"고유값: {unique_values}")
        print("-" * 40)
```

```python
if '공판장코드' in columns and '공판장명' in columns and '품목코드' in columns and '품목명' in columns and '품종코드' in columns and '품종명' in columns:
        unique_combinations = df[['공판장코드', '공판장명', '품목코드', '품목명', '품종코드', '품종명']].drop_duplicates()
        print("unique pair (공판장코드, 공판장명, 품목코드, 품목명, 품종코드, 품종명):")
        print(unique_combinations)
```

```python
# 출력할 최대 행 수를 None으로 설정하여 제한을 없앰
pd.set_option('display.max_rows', None)

if '공판장코드' in columns and '공판장명' in columns and '품목명' in columns:
    unique_combinations = df[['공판장코드', '공판장명', '품목명']].drop_duplicates()
    print("unique pair (공판장코드, 공판장명, 품목명):")
    print(unique_combinations)
```
