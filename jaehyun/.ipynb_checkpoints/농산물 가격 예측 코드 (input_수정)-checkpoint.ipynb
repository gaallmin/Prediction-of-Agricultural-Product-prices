```python
def correlation(x, y):
    T = int(len(x))
    
    cor = np.zeros(T)

    
    ux = np.mean(x)
    uy = np.mean(y)
    stdx = np.std(x)
    stdy = np.std(y)
    for t in range(T):
        cor[t] = ((ux - x[t])*(uy - y[t]))
    
    return np.mean(cor)/(stdx*stdy)




corr = []
auto_corr = []

for i in range(10):
    corr_i = []
    auto_corr_i = []
    for j in range(10):
        print('i = {}'.format(train_set[i][0]))
        print('j = {}'.format(train_set[j][0]))
        print('cor = {}'.format(correlation(train_set[i][1:],train_set[j][1:])))
        plt.plot((train_set[i][1:]-np.min(train_set[i][1:]))/(np.max(train_set[i][1:]) - np.min(train_set[i][1:])))
        plt.plot((train_set[j][1:]-np.min(train_set[j][1:]))/(np.max(train_set[j][1:]) - np.min(train_set[j][1:])))
        plt.show()
        corr_i.append(correlation(train_set[i][1:],train_set[j][1:]))
#     auto_corr.append(auto_correlation(train_set[i][1:]))
    corr.append(corr_i)
        
        
```

```python
# 인풋 데이터2로 자기상관계수 (t_lag = 1) 를 고려

def auto_correlation(x):
    T = len(x) - 2
    cor = np.zeros(T)
    std1 = np.std(x)
    ux1 = np.mean(x)

    
    
    for t in range(T):
        cor[t] = ((x[t+1] - x[t])*(x[t+2] - x[t+1]))
    
    if std1 != 0: 
        return np.mean(cor)/(std1*std1)
    
    else:
        return 0
```

```python
# input dimension (9) -> output 1

def data_seperate(data):
    train_x = []
    train_y = []
    auto_corr = []
    data = data[1:]
    T = len(data) - 11
    
    for t in range(T):
        x = []
        for n in range(9):
            x.append(data[t+n])
     
        train_x.append(x)
        train_y.append(data[t+9:t+12])
        
        
    return np.asarray(train_x).reshape(len(train_x),9), np.asarray(train_y).reshape(len(train_y),3)
```
