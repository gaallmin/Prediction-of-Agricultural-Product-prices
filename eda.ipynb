## Data Cleaning

```python
# parsing 시점 column 
# 하순 : 5th, 중순: 15th, 상순: 25th
def parse_date(date):
    year = int(date[:4])
    month = int(date[4:6])
    period = date[6:]

    if period == '하순':
        day = 5
    elif period == '중순':
        day = 15
    elif period == '상순':
        day = 25
    else:
        day = 1
        
    return pd.Timestamp(year=year, month=month, day=day)

df['datetime'] = df['시점'].apply(parse_date)
df_train1['datetime'] = df_train1['시점'].apply(parse_date)
df_train2['datetime'] = df_train2['시점'].apply(parse_date)

df.set_index('datetime', inplace=True)
df_train1.set_index('datetime', inplace=True)
df_train2.set_index('datetime', inplace=True)

print(df.head())
```

# EDA 

### Key Focus Areas for EDA and Modeling
1. Time Series Characteristics:
- Trends: Identify overall direction in the data (upward/downward).
- Seasonality: Detect repeating patterns at regular intervals (monthly, yearly).
- Stationarity: Ensure the series has constant mean and variance over time.

2. Meta Data Utilization:
- Transaction Volume and Amount: Understand how transaction metrics correlate with prices.
- Auction Counts: Explore the impact of auctions on price fluctuations.
- Previous Period Prices: Utilize lagged price information from meta data.

3. Feature Relationships:
- Correlation Analysis: Determine which features are most strongly associated with the target.
- Interaction Effects: Examine how combinations of features affect prices.

4. Handling Categorical Variables:
- Encoding Techniques: Use One-Hot Encoding for tree-based models.
- Consistency: Ensure consistent encoding across training and test datasets.

5. Temporal Integrity:
- Avoid Data Leakage: Ensure that no future information is used to predict past or current prices.
- Validation Strategy: Use time-based splits like TimeSeriesSplit to maintain the sequence.

```python
# Group the data by '품목명' and '품종명' and calculate the mean price for each group
grouped_data = df.groupby(['품목명', '품종명', '시점'])['평균가격(원)'].mean().reset_index()

# Check the head of the grouped data
print(grouped_data.head())
```

```python
import matplotlib.pyplot as plt

# Function to plot trend for each 품목 and 품종
def plot_trend_by_product_variety(df, product_name):
    # Filter data for the selected product
    product_data = df[df['품목명'] == product_name]

    # Get the unique 품종명 (varieties) for the selected 품목
    varieties = product_data['품종명'].unique()

    # Plot price trend for each 품종
    plt.figure(figsize=(10, 6))
    for variety in varieties:
        variety_data = product_data[product_data['품종명'] == variety]
        plt.plot(variety_data['시점'], variety_data['평균가격(원)'], label=variety)

    plt.title(f'Price Trend for 품목: {product_name}')
    plt.xlabel('시점')
    plt.ylabel('평균가격(원)')
    plt.xticks(rotation=45)
    plt.legend(title='품종명')
    plt.show()

plot_trend_by_product_variety(grouped_data, '건고추')
plot_trend_by_product_variety(grouped_data, '배')
```
